# Dependent t-Test ******

Does the Source of the Anime Influence the Type of Anime?
Use the variable source, recoded to have four levels:

Manga
Book
Game
Listening
And use the variable type.

```python
## import packages 
```

```python
import pandas as pd
from scipy import stats
```

```python
anime
```

```python
anime.value_counts("source")
```

```python
def Source1(series):
    if series == "Manga":
        return "Manga"
    if series == "4-koma manga": 
        return "Manga"
    if series == "Digital manga": 
        return "Manga"
    if series == "Web manga": 
        return "Manga"
    if series == "Original":
        return "Book"
    if series == "Light novel":
        return "Book"
    if series == "Visual novel":
        return "Book"
    if series == "Novel":
        return "Book"
    if series == "Picture book":
        return "Book"
    if series == "Book":
        return "Book"
    if series == "Game":
        return "Game"
    if series == "Card game":
        return "Game"
    if series == "Music":
        return "Listening"
    if series == "Radio":
        return "Listening"
    if series == "Other":
        return "Listening"

anime["SourceR"] = anime["source"].apply(Source1)
```

```python
anime.head()
```

## Test Assumptions

```python
anime['SourceR'].hist()
```

```python
anime['type'].hist()
```

## Run the Analysis

```python
stats.ttest_rel(anime['SourceR'],anime['type'])
```
